# DATA PROCESSING 

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import metrics
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler
from sklearn.linear_model import LogisticRegression, LinearRegression, ElasticNet, Lasso, Ridge
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.ensemble import BaggingClassifier, BaggingRegressor,RandomForestClassifier,RandomForestRegressor
from sklearn.svm import LinearSVC, LinearSVR, SVC, SVR
from xgboost import XGBClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
total_transaction = len(data)
normal_transaction = len(data[data.Class == 0])
fraudulent_transaction = len(data[data.Class == 1])
fraud_percentage = round(fraudulent_transaction/normal_transaction*100,2)
print("The total number of transactions : ",total_transaction)
print("The total number of normal transactions : ",normal_transaction)
print("The total number of fraudulent transactions : ",fraudulent_transaction)
print("Fraud Percentage : ",fraud_percentage)
```

```python
#SCALING DOWN THE AMOUNT VARIABLES 
```

```python
sc = StandardScaler()
amount = data['Amount'].values
amount
data['Amount'] = sc.fit_transform(amount.reshape(-1,1))
data['Amount']
```

```python
data.drop(['Time'],axis=1,inplace=True)
```

```python
data.head()
```

```python
#Removing the duplicates 
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
#Train and test split 
```

```python
X = data.iloc[:,:-1]
Y = data.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.25,random_state=1)
```

# MODEL BUILDING 

# DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=4,criterion='entropy')
model.fit(X_train, Y_train)
predictions = model.predict(X_test) 
```

```python
accuracy = accuracy_score(predictions,Y_test)
```

```python
print("The accuracy of the model is : ",accuracy)
```

```python
score = f1_score(predictions,Y_test)
```

```python
print("The F1 Score of the model is : ",score)
```

```python
matrix = confusion_matrix(predictions, Y_test)
```

```python
print("The Confusion Matrix of the model is : ",matrix)
```
