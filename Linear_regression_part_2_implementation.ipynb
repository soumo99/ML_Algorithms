```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
df = load_boston()
```

```python
df
```

```python
pd.DataFrame(df.data)
```

```python
dataset = pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.head(10)
```

```python
dataset.columns = df.feature_names
```

```python
dataset.head(10)
```

```python
#Independent features and dependent features
X = dataset
Y = df.target
```

```python
Y
```

```python
#Train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train,y_test = train_test_split(X,Y,test_size=0.30,random_state=42)
```

```python
X_train
```

```python
#we will standardize x_test with fit transform and for x_train we will standardize with transform 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit_transform(X_train)
```

```python
X_train =  scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
#Crossvalidation
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=10)
```

```python
mse = cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=10)
```

```python
#Mean calculation 
np.mean(mse)
```

```python
#Predicting the test data 
reg_predict = regression.predict(X_test)
```

```python
reg_predict
```

```python
#Checking the predicted value with the truth value i.e Y_test
import seaborn as sns
sns.displot(reg_predict-y_test, kind='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(reg_predict,y_test)
```
