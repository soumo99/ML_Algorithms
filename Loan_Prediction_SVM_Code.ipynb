```python
#Removing the null values 
data = data.dropna(axis=0)
```

```python
data.isnull().sum()
```

```python
data = data.replace(to_replace = '3+', value = 4 )
```

```python
data.replace({'Married':{'No':0,'Yes':1},'Gender':{'Male':1,'Female':0},'Self_Employed':{'No':0,'Yes':1},
                      'Property_Area':{'Rural':0,'Semiurban':1,'Urban':2},'Education':{'Graduate':1,'Not Graduate':0}},inplace=True)
```

```python
data.head()
```

```python
#Changing data values from string to numeric
data['Dependents'] = pd.to_numeric(data['Dependents'],errors='coerce')
```

```python
#Dividing the dataset between independent and dependent variables 
X = data.drop(['Loan_ID','Loan_Status'],axis=1)
Y = data.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
#Splitting the dataset between training and testing 
from sklearn.model_selection import train_test_split
X_train,X_test ,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
X_train
```

```python
Y_train
```

```python
#Using Standardization technique to scale down the values 
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel = 'linear')
```

```python
model.fit(X_train,Y_train)
```

```python
#Checking prediction rate for training data 
predict_train = model.predict(X_train)
```

```python
from sklearn.metrics import classification_report
cr_train = classification_report(predict_train,Y_train)
```

```python
print(cr_train)
```

```python
#Checking prediction rate for testing data 
predict_test = model.predict(X_test)
```

```python
cr_test = classification_report(predict_test,Y_test)
```

```python
print(cr_test)
```

# Using Hyperparameters tunning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1,1,10,11,100,1000],
             'gamma':[1,0.1,0.01,0.001,0.0001],
             'kernel': ['rbf']}

#verbose just means the text output describing the process
grid_model = GridSearchCV(SVC(),param_grid,refit = True,verbose = 3)
```

```python
grid_model.fit(X_train,Y_train)
```

```python
#checking the best parameters after tuning 
print(grid_model.best_params_)
```

```python
#How the mold looks after tuning 
print(grid_model.best_estimator_)
```

```python
#Checking prediction rate for training data 
grid_predict_train = grid_model.predict(X_train)
```

```python
grid_cr_train = classification_report(grid_predict_train,Y_train)
```

```python
print(grid_cr_train)
```

```python
#Checking prediction rate for testing data 
grid_predict_test = grid_model.predict(X_test)
```

```python
grid_cr_test = classification_report(grid_predict_test,Y_test)
```

```python
print(grid_cr_test)
```
