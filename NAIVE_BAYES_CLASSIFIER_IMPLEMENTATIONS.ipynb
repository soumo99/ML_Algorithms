```python
#Encoding the quality column 
data['quality'] = data['quality'].map({'bad':0,'good':1})
```

```python
#Dividng the dataset between independent and dependent variables 
X = data.iloc[:,:-1]
Y = data.iloc[:,-1]
```

```python
X
```

```python
#Splitting the dataset between training and testing 
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
#Standdardization process is being used to scale down the values 
from sklearn.preprocessing import StandardScaler
```

```python
X_train_stand = X_train.copy()
X_test_stand = X_test.copy()

cols = ['fixed acidity','free sulfur dioxide','total sulfur dioxide','pH','alcohol']

for x in cols:
    scale = StandardScaler().fit(X_train_stand[[x]])
    
    X_train_stand[x] = scale.transform(X_train_stand[[x]])
    X_test_stand[x] = scale.transform(X_test_stand[[x]])
```

```python
X_train
```

```python
X_train_stand
```

```python
X_test_stand
```

```python
#Fitting the model 
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
```

```python
model.fit(X_train_stand,Y_train)
```

```python
#Predictions 
predictions = model.predict(X_test_stand)
```

```python
predictions
```

```python
#Evaluations/Performance Metrics 

from sklearn.metrics import confusion_matrix , accuracy_score, classification_report
cm = confusion_matrix(Y_test,predictions)
score = accuracy_score(Y_test,predictions)
cr = classification_report(Y_test,predictions)
```
