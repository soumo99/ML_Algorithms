```python
#Merging the 2 datasets 
#on is used to which column is to be combined
#how is used on which dataframe it is to be combined 
pd.merge(data,df , on='Country Code',how = 'left')
```

```python
#saving under a final dataframe

final_data = pd.merge(data,df , on='Country Code',how = 'left')
```

```python
final_data.head(10)
```

```python
#checking the data types
final_data.dtypes
```

```python
#checking the number of records each country holds 
final_data.Country.value_counts()
```

Observation :
It is observed that in India the max number of transactions are happening wrt other countries

```python
country_names = final_data.Country.value_counts().index
```

```python
country_values = final_data.Country.value_counts().values
```

```python
#Ploting Pie chart - Plotting all the country data 
plt.figure(figsize = (20,10))
plt.pie(country_values,labels = country_names)
```

```python
#Ploting Pie chart - Plotting top 3 country data 
plt.figure(figsize = (20,10))
plt.pie(country_values[:3],labels = country_names[:3], autopct='%1.2f%%')
```

Observation - Zomato maximum transactions are from India followed by USA and UK

```python
final_data.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
#creating a dataframe out of the 3 columns
final_data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0 : 'Rating count'})
```

Observations
1. When rating is between 4.5 - 4.9 -----> Excellent
2. When rating is between 4.0 - 3.4 -----> Very Good
3. When rating is between 3.5 - 3.9 -----> Good
4. When rating is between 3.0 - 3.4 -----> Average
5. When rating is between 2.5 - 2.9 -----> Average
6. When rating is between 2.0 - 2.4 -----> Poor

```python
rating_data = final_data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0 : 'Rating count'})
```

```python
rating_data.head()
```

```python
#Visualizations 
plt.figure(figsize = (20,10))
sns.barplot(x = 'Aggregate rating',y = 'Rating count',data = rating_data)
```

```python
plt.figure(figsize = (20,10))
sns.barplot(x = 'Aggregate rating',y = 'Rating count', hue = 'Rating color',data = rating_data, palette = ['blue','red','orange','yellow','green','green'])
```

Observations : 
1. Not rated count is very much high
2. Maximum number of ratings are between 2.5 - 3.4

```python
#Count Plot
plt.figure(figsize = (20,10))
sns.countplot(x = 'Rating color',data = rating_data,palette = ['blue','red','orange','yellow','green','green'])
```

```python
#Finding countries that has 0 ratings 
final_data[final_data['Rating color']=='White'].groupby('Country').size().reset_index()
```

Observations : 
Maximum number of 0  ratings are from Indian Customers 

```python
# Find out which currency is used by which Country 
final_data[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```

```python
#Find out whether countries has online delivery option or not 
final_data[['Country','Has Online delivery']].groupby(['Country','Has Online delivery']).size().reset_index()
```

```python
#Find out which country has online delivery option
final_data[final_data['Has Online delivery'] == 'Yes'].Country.value_counts()
```

Observations :
