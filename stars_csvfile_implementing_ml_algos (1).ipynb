```python
#Removing B,A,F ,K ,G from the spectral_class 

indexes=data[(data['Spectral_Class']=='A') | (data['Spectral_Class'] =='B')| (data['Spectral_Class'] =='F')|(data['Spectral_Class'] =='G')|(data['Spectral_Class'] =='K')].index
data.drop(indexes,inplace=True)
```

```python
#Encoding M and O into 1 and 2
data['Spectral_Class'] = data['Spectral_Class'].map({'M':1,'O':2})
```

```python
data.drop("Color", axis=1, inplace=True)
```

```python
#Differentiating between dependent and independent variables 
X = data.iloc[:,:-2]
Y = data.iloc[:,-2]
```

```python
X
```

```python
Y
```

```python
#Dividing into train test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=4)
```

```python
X_train
```

```python
X_test
```

```python
#Model Creation
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
#Training the data and making predictions
```

```python
model.fit(X_train,y_train)
predictions = model.predict(X_test)
```

```python
#Calculating the performanace metrics
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))

```
